# Loan - Random Forest Project

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('loan_data.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

## EDA

```python
plt.figure(figsize=(10,6))
data[data['credit.policy']==1]['fico'].hist(bins=35,color='blue',alpha=0.6,label='credit policy = 1')
data[data['credit.policy']==0]['fico'].hist(bins=35,color='red',alpha=0.6,label='credit policy = 0')
plt.legend()
plt.xlabel('FICO')
```

```python
plt.figure(figsize=(10,6))
data[data['not.fully.paid']==1]['fico'].hist(bins=35,color='blue',alpha=0.6,label='not fully paid = 1')
data[data['not.fully.paid']==0]['fico'].hist(bins=35,color='red',alpha=0.6,label='not fully paid = 0')
plt.legend()
plt.xlabel('FICO')
```

```python
plt.figure(figsize=(11,7))
sns.countplot(x='purpose',hue='not.fully.paid',data=data,palette='Set1')
```

```python
sns.jointplot(x='fico',y='int.rate',data=data)
```

```python
plt.figure(figsize=(11,7))
sns.lmplot(x='fico',y='int.rate',data=data,hue='credit.policy',palette='Set1',col='not.fully.paid')
```
